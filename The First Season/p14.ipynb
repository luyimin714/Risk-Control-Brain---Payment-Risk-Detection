```python
import gc
import pandas as pd
import numpy as np
import os
import xgboost as xgb
import lightgbm as lgb
from sklearn import metrics
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
####################训练集#######################
DATA_PATH = "datasets"
csv_path = os.path.join(DATA_PATH, "atec_anti_fraud_train.csv")
data = pd.read_csv(csv_path)
```

```python
####################训练集#######################
pd.set_option('display.max_columns',2000)
data
```

```python
data.loc[data['label'] == -1, 'label'] = 1
data.loc[data['label'] == 1, 'label'] = 2
data.loc[data['label'] == 0, 'label'] = 1
data.loc[data['label'] == 2, 'label'] = 0
data['label'].value_counts()
```

```python
####################测试集#######################
csv_path = os.path.join(DATA_PATH, "atec_anti_fraud_test_b.csv")
data_test = pd.read_csv(csv_path)
```

```python
miiss_80 = ['f36', 'f37', 'f38', 'f39', 'f40', 'f41', 'f42', 'f43', 'f44', 'f45', 'f46', 'f47']
data.drop('f36',axis=1, inplace=True)
data.drop('f37',axis=1, inplace=True)
data.drop('f38',axis=1, inplace=True)
data.drop('f39',axis=1, inplace=True)
data.drop('f40',axis=1, inplace=True)
data.drop('f41',axis=1, inplace=True)
data.drop('f42',axis=1, inplace=True)
data.drop('f43',axis=1, inplace=True)
data.drop('f44',axis=1, inplace=True)
data.drop('f45',axis=1, inplace=True)
data.drop('f46',axis=1, inplace=True)
data.drop('f47',axis=1, inplace=True)

data_test.drop('f36',axis=1, inplace=True)
data_test.drop('f37',axis=1, inplace=True)
data_test.drop('f38',axis=1, inplace=True)
data_test.drop('f39',axis=1, inplace=True)
data_test.drop('f40',axis=1, inplace=True)
data_test.drop('f41',axis=1, inplace=True)
data_test.drop('f42',axis=1, inplace=True)
data_test.drop('f43',axis=1, inplace=True)
data_test.drop('f44',axis=1, inplace=True)
data_test.drop('f45',axis=1, inplace=True)
data_test.drop('f46',axis=1, inplace=True)
data_test.drop('f47',axis=1, inplace=True)


data.fillna(0, inplace=True)
data_test.fillna(0, inplace=True)
```

```python
data_train = data[data['date'] <= 20171031]
data_val = data[data['date'] > 20171031]
```

```python
X_train = data_train.loc[:,'f1':].as_matrix()
y_train = data_train.loc[:,'label'].as_matrix()

X_val = data_val.loc[:,'f1':].as_matrix()
y_val = data_val.loc[:,'label'].as_matrix()

X_test = data_test.loc[:,'f1':].as_matrix()
print(X_train.shape)
print(y_train.shape)
print(X_val.shape)
print(y_val.shape)
print(X_test.shape)
```

```python
del data, data_train, data_val
gc.collect()
```

```python
clf = lgb.LGBMClassifier(   
                            boosting_type='gbdt',
                            num_leaves=50,
                            n_estimators=300,
                            objective='binary',
                            learning_rate=0.05,
                            feature_fraction=0.9,
                            bagging_fraction=0.8,
                            random_state=2018,
                            bagging_freq=5,
                            verbose=0,
                            n_jobs=-1     
                        )
clf.fit(X_train, y_train, eval_set=[(X_val, y_val), (X_train, y_train)], eval_metric={'l2', 'auc'}, early_stopping_rounds=50)
```

```python
y_pred = clf.predict_proba(X_test)[:,0]
result = pd.DataFrame({'id':data_test['id'].as_matrix(), 'score':y_pred})
result.to_csv("result5.csv", index=False)
```
