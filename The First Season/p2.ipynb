```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
import lightgbm as lgb
from sklearn.model_selection import train_test_split
```

```python
####################训练数据#######################
DATA_PATH = "datasets"
csv_path = os.path.join(DATA_PATH, "atec_anti_fraud_train.csv")
data = pd.read_csv(csv_path)
```

```python
pd.set_option('display.max_columns',2000)
data
```

```python
#########缺失值填充###############
data.loc[:, 'f1':'f4'] = data.loc[:, 'f1':'f4'].fillna(method='ffill')

data['f5'].fillna(data['f5'].mean(), inplace=True)

data.loc[:, 'f6':'f35'] = data.loc[:, 'f6':'f35'].fillna(method='ffill')

data.loc[:, 'f36':'f47'] = data.loc[:, 'f36':'f47'].fillna(method='backfill')
data.loc[:, 'f36':'f47'] = data.loc[:, 'f36':'f47'].fillna(method='ffill')

data.loc[:, 'f48':'f81'] = data.loc[:, 'f48':'f81'].fillna(method='ffill')

data['f82'].fillna(data['f82'].mean(), inplace=True)
data['f83'].fillna(data['f83'].mean(), inplace=True)
data['f84'].fillna(data['f84'].mean(), inplace=True)
data['f85'].fillna(data['f85'].mean(), inplace=True)
data['f86'].fillna(data['f86'].mean(), inplace=True)

data.loc[:, 'f87':'f297'] = data.loc[:, 'f87':'f297'].fillna(method='ffill')
```

```python
data['label'].value_counts()
```

```python
###########将-1作为有风险#################
data.loc[data['label'] == -1, 'label'] = 1
data['label'].value_counts()
```

```python
############用于预测的数据#################
csv_path = os.path.join(DATA_PATH, "atec_anti_fraud_test_a.csv")
test_data = pd.read_csv(csv_path)
```

```python
test_data
```

```python
#########缺失值填充###############
test_data.loc[:, 'f1':'f4'] = test_data.loc[:, 'f1':'f4'].fillna(method='ffill')
test_data['f5'].fillna(data['f5'].mean(), inplace=True)

test_data.loc[:, 'f6':'f35'] = test_data.loc[:, 'f6':'f35'].fillna(method='ffill')

test_data.loc[:, 'f36':'f47'] = test_data.loc[:, 'f36':'f47'].fillna(method='backfill')
test_data.loc[:, 'f36':'f47'] = test_data.loc[:, 'f36':'f47'].fillna(method='ffill')

test_data.loc[:, 'f48':'f63'] = test_data.loc[:, 'f48':'f63'].fillna(method='ffill')

test_data.loc[:, 'f64':'f81'] = test_data.loc[:, 'f64':'f81'].fillna(method='backfill')
test_data.loc[:, 'f64':'f81'] = test_data.loc[:, 'f36':'f81'].fillna(method='ffill')

test_data['f82'].fillna(test_data['f82'].mean(), inplace=True)
test_data['f83'].fillna(test_data['f83'].mean(), inplace=True)
test_data['f84'].fillna(test_data['f84'].mean(), inplace=True)
test_data['f85'].fillna(test_data['f85'].mean(), inplace=True)
test_data['f86'].fillna(test_data['f86'].mean(), inplace=True)

test_data.loc[:, 'f87':'f160'] = test_data.loc[:, 'f87':'f160'].fillna(method='backfill')
test_data.loc[:, 'f87':'f160'] = test_data.loc[:, 'f87':'f160'].fillna(method='ffill')

test_data.loc[:, 'f161':'f297'] = test_data.loc[:, 'f161':'f297'].fillna(method='ffill')
```

```python
###########训练数据##################
X_df = data.loc[:,'date':'f297']
y_df = data.loc[:,'label']
X = X_df.as_matrix()
y = y_df.as_matrix()
############用于预测的数据###############
X_df_pred = test_data.loc[:,'date':'f297']
X_pred = X_df_pred.as_matrix()
```

```python
####################模型训练##########################
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
print(X_train.shape)
print(X_test.shape)
print(X_pred.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
lgb_train = lgb.Dataset(X_train, y_train)
lgb_eval = lgb.Dataset(X_test, y_test, reference=lgb_train)
params = {
    'task': 'train',
    'boosting_type': 'gbdt',
    'objective': 'binary',
    'metric': {'l2', 'auc'},
    'num_leaves': 31,
    'learning_rate': 0.05,
    'feature_fraction': 0.9,
    'bagging_fraction': 0.8,
    'bagging_freq': 5,
    'verbose': 0
}
```

```python
gbm = lgb.train(params,
                lgb_train,
                num_boost_round=300,
                valid_sets=lgb_eval,
                early_stopping_rounds=5)
```

```python
y_pred = gbm.predict(X_pred, num_iteration=gbm.best_iteration)
```

```python
result = pd.DataFrame({'id':test_data['id'].as_matrix(), 'score':y_pred})
result.to_csv("result3.csv", index=False)
```
