```python
import gc
import pandas as pd
import numpy as np
import os
import xgboost as xgb
import lightgbm as lgb
from sklearn import metrics
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.model_selection import train_test_split
```

```python
####################训练集#######################
DATA_PATH = "datasets"
csv_path = os.path.join(DATA_PATH, "atec_anti_fraud_train.csv")
data = pd.read_csv(csv_path)
```

```python
####################训练集#######################
pd.set_option('display.max_columns',2000)
data
```

```python
data.loc[data['label'] == -1, 'label'] = 1
data.loc[data['label'] == 1, 'label'] = 2
data.loc[data['label'] == 0, 'label'] = 1
data.loc[data['label'] == 2, 'label'] = 0
data['label'].value_counts()
```

```python
####################测试集#######################
csv_path = os.path.join(DATA_PATH, "atec_anti_fraud_test_b.csv")
data_test = pd.read_csv(csv_path)
```

```python
####################测试集#######################
data_test
```

```python
data.fillna(0, inplace=True)
data_test.fillna(0, inplace=True)
```

```python
data_train = data[data['date'] <= 20171031]
data_val = data[data['date'] > 20171031]
```

```python
del data
gc.collect()
```

```python
data_train.shape
```

```python
data_train['date'].value_counts()
```

```python
data_train1 = data_train[data_train['date'] > 20170915]
data_train2 = data_train[(data_train['date'] <= 20170915) | (data_train['date'] > 20170927)]
data_train3 = data_train[(data_train['date'] <= 20170927) | (data_train['date'] > 20171009)]
data_train4 = data_train[(data_train['date'] <= 20171009) | (data_train['date'] > 20171020)]
data_train5 = data_train[data_train['date'] <= 20171020]
print(data_train1.shape)
print(data_train2.shape)
print(data_train3.shape)
print(data_train4.shape)
print(data_train5.shape)
```

```python
del data_train
gc.collect()
```

```python
X_train1 = data_train1.loc[:,'f1':].as_matrix()
y_train1 = data_train1.loc[:,'label'].as_matrix()

X_train2 = data_train2.loc[:,'f1':].as_matrix()
y_train2 = data_train2.loc[:,'label'].as_matrix()

X_train3 = data_train3.loc[:,'f1':].as_matrix()
y_train3 = data_train3.loc[:,'label'].as_matrix()

X_train4 = data_train4.loc[:,'f1':].as_matrix()
y_train4 = data_train4.loc[:,'label'].as_matrix()

X_train5 = data_train5.loc[:,'f1':].as_matrix()
y_train5 = data_train5.loc[:,'label'].as_matrix()

X_val = data_val.loc[:,'f1':].as_matrix()
y_val = data_val.loc[:,'label'].as_matrix()

X_test = data_test.loc[:,'f1':].as_matrix()

print(X_train1.shape, y_train1.shape)
print(X_train2.shape, y_train2.shape)
print(X_train3.shape, y_train3.shape)
print(X_train4.shape, y_train4.shape)
print(X_train5.shape, y_train5.shape)
```

```python
del data_train1, data_train2, data_train3, data_train4, data_train5
gc.collect()
```

```python
clf1 = lgb.LGBMClassifier(   
                            boosting_type='gbdt',
                            num_leaves=50,
                            n_estimators=300,
                            objective='binary',
                            learning_rate=0.05,
                            feature_fraction=0.9,
                            bagging_fraction=0.8,
                            random_state=1,
                            bagging_freq=5,
                            verbose=0,
                            n_jobs=-1     
                        )
clf1.fit(X_train1, y_train1, eval_set=[(X_val, y_val), (X_train1, y_train1)], eval_metric={'l2', 'auc'}, early_stopping_rounds=50)
```

```python
clf2 = lgb.LGBMClassifier(   
                            boosting_type='gbdt',
                            num_leaves=50,
                            n_estimators=300,
                            objective='binary',
                            learning_rate=0.05,
                            feature_fraction=0.9,
                            bagging_fraction=0.8,
                            random_state=2,
                            bagging_freq=5,
                            verbose=0,
                            n_jobs=-1     
                        )
clf2.fit(X_train2, y_train2, eval_set=[(X_val, y_val), (X_train2, y_train2)], eval_metric={'l2', 'auc'}, early_stopping_rounds=50)
```

```python
clf3 = lgb.LGBMClassifier(   
                            boosting_type='gbdt',
                            num_leaves=50,
                            n_estimators=300,
                            objective='binary',
                            learning_rate=0.05,
                            feature_fraction=0.9,
                            bagging_fraction=0.8,
                            random_state=3,
                            bagging_freq=5,
                            verbose=0,
                            n_jobs=-1     
                        )
clf3.fit(X_train3, y_train3, eval_set=[(X_val, y_val), (X_train3, y_train3)], eval_metric={'l2', 'auc'}, early_stopping_rounds=50)
```

```python
clf4 = lgb.LGBMClassifier(   
                            boosting_type='gbdt',
                            num_leaves=50,
                            n_estimators=300,
                            objective='binary',
                            learning_rate=0.05,
                            feature_fraction=0.9,
                            bagging_fraction=0.8,
                            random_state=4,
                            bagging_freq=5,
                            verbose=0,
                            n_jobs=-1     
                        )
clf4.fit(X_train4, y_train4, eval_set=[(X_val, y_val), (X_train4, y_train4)], eval_metric={'l2', 'auc'}, early_stopping_rounds=50)
```

```python
clf5 = lgb.LGBMClassifier(   
                            boosting_type='gbdt',
                            num_leaves=50,
                            n_estimators=300,
                            objective='binary',
                            learning_rate=0.05,
                            feature_fraction=0.9,
                            bagging_fraction=0.8,
                            random_state=5,
                            bagging_freq=5,
                            verbose=0,
                            n_jobs=-1     
                        )
clf5.fit(X_train5, y_train5, eval_set=[(X_val, y_val), (X_train5, y_train5)], eval_metric={'l2', 'auc'}, early_stopping_rounds=50)
```

```python
y_pred1 = clf1.predict_proba(X_test)[:,0]
y_pred2 = clf2.predict_proba(X_test)[:,0]
y_pred3 = clf3.predict_proba(X_test)[:,0]
y_pred4 = clf4.predict_proba(X_test)[:,0]
y_pred5 = clf5.predict_proba(X_test)[:,0]
```

```python
print(y_pred1[:10])
print(y_pred2[:10])
print(y_pred3[:10])
print(y_pred4[:10])
print(y_pred5[:10])
```

```python
y_pred = y_pred1 * 0.2 + y_pred2 * 0.2 + y_pred3 * 0.2 + y_pred4 * 0.2 + y_pred5 * 0.2
print(y_pred[:10])
```

```python
result = pd.DataFrame({'id':data_test['id'].as_matrix(), 'score':y_pred})
result.to_csv("result4.csv", index=False)
```
