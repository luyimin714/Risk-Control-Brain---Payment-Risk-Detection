```python
import pandas as pd
import numpy as np
import os
import xgboost as xgb
import lightgbm as lgb
from sklearn import metrics
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
```

```python
####################训练集#######################
DATA_PATH = "datasets"
csv_path = os.path.join(DATA_PATH, "atec_anti_fraud_train.csv")
data = pd.read_csv(csv_path)
```

```python
####################训练集#######################
pd.set_option('display.max_columns',2000)
data
```

```python
data['label'].value_counts()
```

```python
data.loc[(data['label'] == -1) & (data['f57'] > 110), 'label'] = 0
data.loc[(data['label'] == -1) & (data['f96'] > 10), 'label'] = 0
data.loc[(data['label'] == -1) & (data['f105'] > 225), 'label'] = 0
data.loc[(data['label'] == -1) & (data['f234'] > 120), 'label'] = 0
data.loc[(data['label'] == -1) & (data['f206'] > 120), 'label'] = 0
data.loc[(data['label'] == -1) & (data['f210'] > 250), 'label'] = 0
```

```python
data['label'].value_counts()
```

```python
data.loc[data['label'] == -1, 'label'] = 1
data['label'].value_counts()
```

```python
####################训练集(填充缺失值)#######################
data.fillna(0, inplace=True)
```

```python
####################训练集(将年月日分开)#######################
data['year'] = data['date'].apply(lambda x: int(str(x)[0:4]))
data['month'] = data['date'].apply(lambda x: int(str(x)[4:6]))
data['day'] = data['date'].apply(lambda x: int(str(x)[6:8]))
```

```python
####################测试集#######################
csv_path = os.path.join(DATA_PATH, "atec_anti_fraud_test_a.csv")
data_test = pd.read_csv(csv_path)
```

```python
####################测试集#######################
data_test
```

```python
####################测试集(填充缺失值)#######################
data_test.fillna(0, inplace=True)
```

```python
#################### 测试集(将年月日分开) #######################
data_test['year'] = data_test['date'].apply(lambda x: int(str(x)[0:4]))
data_test['month'] = data_test['date'].apply(lambda x: int(str(x)[4:6]))
data_test['day'] = data_test['date'].apply(lambda x: int(str(x)[6:8]))
```

```python
#################### 数据划分 #######################
data_train = data[data['month'] != 11]
data_val = data[data['month'] == 11]
```

```python
data_train.drop(['date', 'year', 'month'], axis=1, inplace=True)
data_val.drop(['date', 'year', 'month'], axis=1, inplace=True)
```

```python
data_train
```

```python
data_val
```

```python
data_test.drop(['date', 'year', 'month'], axis=1, inplace=True)
```

```python
#################### 数据划分 #######################
X_train = data_train.loc[:, 'f1':].as_matrix()
y_train = data_train.loc[:, 'label'].as_matrix()
X_val = data_val.loc[:, 'f1':].as_matrix()
y_val = data_val.loc[:, 'label'].as_matrix()
X_test = data_test.loc[:, 'f1':].as_matrix()
```

```python
def eval_func(y_pred, y_true):
    fpr, tpr, thresholds = metrics.roc_curve(y_pred, y_true, pos_label=1)
    score = 0.4 * tpr[np.argwhere(fpr < 0.001)[-1]] + 0.3 * tpr[np.argwhere(
        fpr < 0.005)[-1]] + 0.3 * tpr[np.argwhere(fpr < 0.01)[-1]]
    return 'score', score, True
```

```python
clf = lgb.LGBMClassifier(   
                            boosting_type='gbdt',
                            num_leaves=60,
                            n_estimators=200,
                            objective='binary',
                            learning_rate=0.05,
                            feature_fraction=0.9,
                            bagging_fraction=0.8,
                            bagging_freq=5,
                            random_state=2018,
                            verbose=0,
                            n_jobs=-1     
                        )
clf.fit(X_train, y_train, eval_set=[(X_val, y_val), (X_train, y_train)], eval_metric=eval_func,early_stopping_rounds=50)
```

```python
y_pred = clf.predict_proba(X_test)[:,1]
result = pd.DataFrame({'id':data_test['id'].as_matrix(), 'score':y_pred})
result.to_csv("result15.csv", index=False)
```
