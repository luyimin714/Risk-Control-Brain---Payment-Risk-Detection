```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
import lightgbm as lgb
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import Imputer
```

```python
####################训练数据#######################
DATA_PATH = "datasets"
csv_path = os.path.join(DATA_PATH, "atec_anti_fraud_train.csv")
data = pd.read_csv(csv_path)
```

```python
pd.set_option('display.max_columns',2000)
data
```

```python
#########缺失值填充###############
### f5 f82 f83 f84 f85 f86使用均值填充#############
mean_f5 = data['f5'].mean()
mean_f82 = data['f82'].mean()
mean_f83 = data['f83'].mean()
mean_f84 = data['f84'].mean()
mean_f85 = data['f85'].mean()
mean_f86 = data['f86'].mean()

data['f5'].fillna(mean_f5, inplace=True)
data['f82'].fillna(mean_f82, inplace=True)
data['f83'].fillna(mean_f83, inplace=True)
data['f84'].fillna(mean_f84, inplace=True)
data['f85'].fillna(mean_f85, inplace=True)
data['f86'].fillna(mean_f86, inplace=True)
```

```python
#########缺失值填充###############
###其余各列使用中值填充#############
column = [0, 1, 2]
data_drop = data.drop(data.columns[column], axis=1) 
imputer = Imputer(strategy="median")
imputer.fit(data_drop)
data_numpy = imputer.transform(data_drop)
data_tr = pd.DataFrame(data_numpy, columns=data.columns[3:])
data.loc[:, 'f1':'f297'] = data_tr
```

```python
data.loc[data['label'] == -1, 'label'] = 1
data['label'].value_counts()
```

```python
############用于预测的数据#################
csv_path = os.path.join(DATA_PATH, "atec_anti_fraud_test_a.csv")
test_data = pd.read_csv(csv_path)
```

```python
pd.set_option('display.max_columns',2000)
test_data
```

```python
#########缺失值填充###############
### f5 f82 f83 f84 f85 f86使用均值填充#############
test_data['f5'].fillna(mean_f5, inplace=True)
test_data['f82'].fillna(mean_f82, inplace=True)
test_data['f83'].fillna(mean_f83, inplace=True)
test_data['f84'].fillna(mean_f84, inplace=True)
test_data['f85'].fillna(mean_f85, inplace=True)
test_data['f86'].fillna(mean_f86, inplace=True)
```

```python
#########缺失值填充###############
###其余各列使用中值填充#############
column = [0, 1]
test_data_drop = test_data.drop(test_data.columns[column], axis=1) 
imputer = Imputer(strategy="median")
imputer.fit(test_data_drop)
test_data_numpy = imputer.transform(test_data_drop)
test_data_tr = pd.DataFrame(test_data_numpy, columns=test_data.columns[2:])
test_data.loc[:, 'f1':'f297'] = test_data_tr
```

```python
###########训练数据##################
X_df = data.loc[:,'date':'f297']
y_df = data.loc[:,'label']
X = X_df.as_matrix()
y = y_df.as_matrix()
############用于预测的数据###############
X_df_pred = test_data.loc[:,'date':'f297']
X_pred = X_df_pred.as_matrix()
####################模型训练##########################
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
lgb_train = lgb.Dataset(X_train, y_train)
lgb_eval = lgb.Dataset(X_test, y_test, reference=lgb_train)
params = {
    'task': 'train',
    'boosting_type': 'gbdt',
    'objective': 'binary',
    'metric': {'l2', 'auc'},
    'num_leaves': 31,
    'learning_rate': 0.05,
    'feature_fraction': 0.9,
    'bagging_fraction': 0.8,
    'bagging_freq': 5,
    'verbose': 0
}
```

```python
gbm = lgb.train(params,
                lgb_train,
                num_boost_round=300,
                valid_sets=lgb_eval,
                early_stopping_rounds=5)
```

```python
y_pred = gbm.predict(X_pred, num_iteration=gbm.best_iteration)
```

```python
result = pd.DataFrame({'id':test_data['id'].as_matrix(), 'score':y_pred})
result.to_csv("result4.csv", index=False)
```
